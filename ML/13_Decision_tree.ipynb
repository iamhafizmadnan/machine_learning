# Decision Tree
Decision tree is a supervised learning algorithm that can be used for both classification and regression tasks. It works by recursively splitting the data into subsets based on the feature that provides the best separation of the target variable. The resulting tree structure can be easily visualized and interpreted, making it a popular choice for many machine learning problems.



```python
import math
```

```python
# example dataset
n_a = 5
n_b = 10    
total = n_a + n_b
```

```python
# lets culculate the proportion
n_a = n_a / total
n_b = n_b / total
print('Proportion of class A:', n_a)
print('Proportion of class B:', n_b)
```

```python
# entropy calculate 
entropy = - (n_a * math.log2(n_a) + n_b * math.log2(n_b))
print('Entropy:', entropy)
```
