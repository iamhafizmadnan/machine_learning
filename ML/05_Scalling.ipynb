Standard Scalling is a feature-scaling technique (also called **z-score normalization**) that transforms numerical features so they have:

- **Mean = 0**
- **Standard Deviation = 1**

This helps models perform better when features are on different scales, especially for **KNN, SVM, PCA, Logistic Regression,** and **Neural Networks**.

Formula: `z = (x - μ) / σ`

```python
# import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler 
```

```python
df = {
    "age": [25, 30, 35, 40, 45],
    "income": [50000, 60000, 70000, 80000, 90000],
    "height": [160, 165, 170, 175, 180]
}

df = pd.DataFrame(df)
df
```

```python
# import the scaler
scaler = StandardScaler()
# fit the scaler to the data
scaler.fit(df)

scaled_data = scaler.transform(df)
scaled_data
```

# Min -Max-Scaler

```python
scaler = MinMaxScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)      
scaled_data
```

# Max- Absolute-Scaler

```python
scaler = MaxAbsScaler()
scaler.fit(df)  
scaled_data = scaler.transform(df)
scaled_data
```

 # Transformation

```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
df = np.random.exponential(scale=2, size=1000)
df = pd.DataFrame(df, columns=["value"])
df.head()
```

```python
df.shape
```

```python
sns.histplot(df["value"], bins=30)
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer

pt_boxcox = PowerTransformer(method="box-cox")
pt_yeo_johnson = PowerTransformer(method="yeo-johnson")
pt_quantile = QuantileTransformer(output_distribution="normal")

df['value_boxcox'] = pt_boxcox.fit_transform(df[["value"]] + 1)
df['value_yeo_johnson'] = pt_yeo_johnson.fit_transform(df[["value"]])
df['value_quantile'] = pt_quantile.fit_transform(df[["value"]])

```

```python
df.head()
```

```python
for col in df.columns:
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(col)
    plt.show()
```

#   Normalization

```python
from sklearn.preprocessing import Normalizer
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
normalizer = Normalizer(norm = "l2")
print(normalizer.fit_transform(data))
```
