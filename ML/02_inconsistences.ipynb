```python
import pandas as pd
```

```python

# from math import nan


data = {
    'date': ['2021-01-01', '01-02-2021', '2021/03/01', '2021.04.01'],
    'name': ['Alice da', 'alice da', 'Alice Da', 'Alice Da'],
    'country': ['USA', 'usa', 'Usa', 'USA'],
    'sales_2020': [100, None, 150, 200],
    'sales_2021': [110, 120, None, 210]
}


df = pd.DataFrame(data)

```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.head()
```

```python
# complete fill the date
df['date'] = df['date'].fillna(pd.Timestamp('2021-01-01'))
df.head()
```

```python
df['name'] =  df['name'].replace({'Alice da':'Alice Da', 'alice da':'Alice Da' })
df.head()

```

```python
df['country'] = df['country'].replace({'usa':'USA', 'Usa':'USA'})
df.head()
```
