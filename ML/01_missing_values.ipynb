# Handling Missing Values


```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# check the null value
df.isnull().sum().sort_values(ascending= False)
```

```python
round(df.isnull().sum() / len (df) * 100, 4).sort_values(ascending = False)
```

# There are several method to fill the data
# 1. Mean , Medain, Mode
# 2. Forward and Backward imputation
# 3. Iterative Imputation
# 4. KNN Imputation (Deep Learning Techinque)


```python
 # missing age data with mean
 df['age'] = df['age'].fillna(df['age'].median())
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
# drop the deck cloumns
df.drop('deck' , axis = 1, inplace = True)
```

```python
df['embarked'].value_counts()
```

```python
# replacing embarked missing values with with mode
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df.isnull().sum()
```

## we can impute using sklearn


```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.impute import SimpleImputer
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# impute the data using simple imputer
imputer= SimpleImputer(strategy = 'median')
df['age'] = imputer.fit_transform(df[['age']])
 
```

```python
df.isnull().sum().sort_values(ascending = False)
```

# Filling value using multivariate

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imputer = IterativeImputer()
df['age'] = imputer.fit_transform(df[['age']])

```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# ffill and bfill
df['age'] = df['age'].ffill()
```

```python
df['age'] = df['age'].bfill()
```
