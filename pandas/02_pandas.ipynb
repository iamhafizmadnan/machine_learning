# 03- how to remane columns in pandas


```python
# 03- how to remane columns in pandas
df.rename(columns={"a": "A", "b": "B", "c": "C"}, inplace=True)
```

```python
df
```

```python
#rename the columns in pandas
df.columns=['Aa' , 'Bb', 'Cc']
df 
```

```python
# replace any character in the column name with another character
df.columns = df.columns.str.replace('a', 'A').str.replace('b', 'B').str.replace('c', 'C')
df
```

```python
df.replace('a', 'A', regex=True).replace('b', 'B', regex=True).replace('c', 'C', regex=True)
```

```python
df
```

```python
df.rename(columns ={'Aa_Aa_': 'A', 'Bb_Bb_': 'B', 'Cc_Cc_': 'C'}, inplace=True)
```

```python
df
```

```python
#add prefix to column names in pandas
df = df.add_prefix('A')
```
